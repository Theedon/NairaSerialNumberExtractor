```python
import pickle
import cv2
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")

store= pickle.load(open("/content/drive/MyDrive/Ogunleye Project/CODES/pickled files/store.pkl", "rb"))

def note_count():
  print("The number of 5 naira notes are: ", len(store[0]))
  print("The number of 10 naira notes are: ", len(store[1]))
  print("The number of 20 naira notes are: ", len(store[2]))
  print("The number of 50 naira notes are: ", len(store[3]))
  print("The number of 100 naira notes are: ", len(store[4]))
  print("The number of 200 naira notes are: ", len(store[5]))
  print("The number of 500 naira notes are: ", len(store[6]))
  print("The number of 1000 naira notes are: ", len(store[7]))

  total= 0
  for i in store:
    total+= len(i)

  print("\n\nThe total number of notes is ", total)
  return 0

def show(currency, a):
  
  if currency == "five":
    g= 0
    #crop= img[400:450, 770:1150]
  if currency == "ten":
    g= 1
    #crop= img[400:470, 770:1150]
  if currency == "twenty":
    g= 2
    #crop= img[400:470, 770:1150]
  if currency == "fifty":
    g= 3
    #crop= img[400:470, 770:1150]
  if currency == "hundred":
    g= 4
    #crop= img[100:190, 760:1020]
  if currency == "two-hundred":
    g= 5
    #crop= img[100:190, 750:1000]
  if currency == "five-hundred":
    g= 6
    #crop= img[100:190, 720:1000]
  if currency == "one-thousand":
    g= 7
    #crop= img[30:120, 50:290]



  img= cv2.resize(store[g][a], dsize=(1100, 550), interpolation= cv2.INTER_LINEAR)
 
  #plt.imshow(img)
  if g == 0: 
    crop= img[400:450, 770:1150]
  if g == 1: 
    crop= img[400:470, 770:1150]
  if g == 2: 
    crop= img[400:470, 770:1150]
  if g == 3: 
    crop= img[400:470, 770:1150]
  if g == 4: 
    crop= img[100:190, 760:1020]
  if g == 5: 
    crop= img[100:190, 750:1000]
  if g == 6: 
    crop= img[100:190, 720:1000]
  if g == 7: 
    crop= img[30:120, 50:290]

  #plt.imshow(crop)


  img= cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  bfilter = cv2.bilateralFilter(gray, 11, 17, 17) #Noise reduction
  edged = cv2.Canny(bfilter, 30, 200) #Edge detection

  plt.imshow(edged)

  return 0
```

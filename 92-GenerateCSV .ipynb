```python
import pickle
import pandas as pd
import numpy as np
import os
import cv2
```

```python
filtered_images= pickle.load(open(PATH_TO_DATA+"filtered_final_images.pkl", "rb"))
test_truth= pickle.load(open(PATH_TO_DATA+"test_original.pkl", "rb"))
final_test_pred= pickle.load(open(PATH_TO_DATA+"final_test_pred.pkl", "rb"))
test_pred_easyocr= final_test_pred[0]
test_pred_kerasocr= final_test_pred[1]
test_pred_pytesseract= final_test_pred[2]
```

# Add images to folder

```python
note_list= ["5", "10", "20", "50", "100", "200", "500", "1000"]
```

```python
%cd /content

folder= "/content/csv_and_images/"
if os.path.isdir(folder):
  !rm -rf $folder 

for i in note_list:
  new_folder= folder+i
  !mkdir -p $new_folder 
```

```python
filenames= []

for i in range(len(filtered_images)):
  current_folder= folder+note_list[i]
  %cd $current_folder
  for j in range(len(filtered_images[i])):
    file_name= str(note_list[i])+"-"+str(j+1)+".jpg"
    filenames.append(file_name)
    cv2.imwrite(file_name, filtered_images[i][j])

%cd /content
```

# Create CSV file

```python
easyocr= []
kerasocr= []
pytesseract= []
for i in range(len(test_pred_easyocr)):
  easyocr+= test_pred_easyocr[i]
  kerasocr+= test_pred_kerasocr[i]
  pytesseract+= test_pred_pytesseract[i]
```

```python
data_dict={"Image_name":  filenames, "EasyOCR": easyocr, "KerasOCR": kerasocr, "PyTesseract": pytesseract}
data_frame= pd.DataFrame(data_dict)
```
